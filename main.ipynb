Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Read Data

```python
players = pd.read_csv('data/Players.csv')
teams = pd.read_csv('data/Teams.csv')
```

```python
players.head()
```

```python
teams.head()
```

```python
teams["Earnings"] = teams['Earnings'].str.replace('$', '')
```

```python
players["Earnings"] = players['Earnings'].str.replace('$','')
players['Earnings'] = players['Earnings'].str.replace(',','')
players['Earnings'] = pd.to_numeric(players['Earnings'])
```

```python
teams.head()
```

```python
players.head()
```

```python
players['Total_Medals'] = players['Gold'] + players['Bronze'] + players['Silver']
players.head()
```

```python
top_10_players = players.sort_values('Total_Medals', ascending = False).head(10)
plt.bar(top_10_players['Player'], top_10_players['Total_Medals'].head(10))
plt.title('Top 10 Pro Players With Most Total Medals')
plt.xticks(rotation = 90)
plt.xlabel('Player Names')
plt.ylabel('Total # of Medals')
plt.show()
```

```python
top_10_gold_medal_players = players.sort_values('Gold', ascending = False).head(10)
plt.bar(top_10_gold_medal_players['Player'], top_10_gold_medal_players['Gold'])
plt.title('Players With the Most Total Gold Medals')
plt.xticks(rotation = 90)
plt.xlabel('Players')
plt.ylabel('Number of Gold Medals')
plt.show()
```

```python
top_10_silver_medal_players = players.sort_values('Silver', ascending = False).head(10)
plt.bar(top_10_silver_medal_players['Player'], top_10_silver_medal_players['Silver'])
plt.title('Players With the Most Total Silver Medals')
plt.xticks(rotation = 90)
plt.xlabel('Players')
plt.ylabel('Number of Silver Medals')
plt.show()
```

```python
top_10_bronze_medal_players = players.sort_values('Bronze', ascending = False).head(10)
plt.bar(top_10_bronze_medal_players['Player'], top_10_bronze_medal_players['Bronze'])
plt.title('Players With the Most Total Bronze Medals')
plt.xticks(rotation = 90)
plt.xlabel('Players')
plt.ylabel('Number of Bronze Medals')
plt.show()
```

```python
top_10_player_earnings = players.sort_values('Earnings', ascending = False).head(10)
plt.bar(top_10_player_earnings['Player'], top_10_player_earnings['Earnings'])
plt.title('Players With the most Earnings')
plt.xticks(rotation = 90)
plt.xlabel('Players')
plt.ylabel('Earnings ($)')
plt.show()
```

Correlation HeatMap Between Total Earnings and Total Gold Medals

```python
players_subset = players[['Earnings', 'Total_Medals']]
corr = players_subset.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
plt.show()
```

There is a very small positive correlation between a players total number of medals and their total earnings

```python
players_subset = players[['Earnings', 'Gold']]
corr = players_subset.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
plt.show()
```

There is a slightly larger positive correlation between a players total gold medals and their earnings

```python
players_subset = players[['Earnings', 'Silver']]
corr = players_subset.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
plt.show()
```

There is no correlation between a players total silver medals and earnings

```python
players_subset = players[['Earnings', 'Bronze']]
corr = players_subset.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)
plt.show()
```

There is no correlation between a players total bronze medals and his total earnings

```python
plt.figure(figsize=(8, 6))
sns.heatmap(players.corr(), annot=True)
plt.show()
```

From the heat map above, we can see the number of medals is not what determines the total amount earned but rather if the player is considered an S Tier player
